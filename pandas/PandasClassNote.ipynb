```python
# pandas merging, joining and concatenating
df_customer = pd.DataFrame({
    'id':[1,2,3,4],
    'name':['Tom','Jerry','James','Dan']
    
})
df_info = pd.DataFrame({
    'id':[2,3,4,5],
    'age':[31,20,40,70],
    'sex':['F','M','M','F']
    
})
```

```python
df_customer
```

```python
df_info
```

```python
pd.merge(df_customer,df_info)
```

```python
df_customer.merge(df_info)
```

```python
pd.merge(df_customer,df_info,on='id')  # id is common so it is specified
```

```python
df_order = pd.DataFrame({
    'id': [2,3,4,5],
    'name': ['Jenny', 'James', 'Dan', 'leo'],
    'quantity': [2,4,6,10]
})
df_order
```

```python
pd.merge(df_customer,df_order)
```

```python
# To combine data on multiple common columns, you can pass a list to on:

# with multiple key columns
df_order = pd.DataFrame({
    'id': [2,3,4,5],
    'name': ['Jenny', 'James', 'Dan', 'leo'],
    'quantity': [2,4,6,10]
})

pd.merge(df_customer, df_order, on=['id', 'name'])
```

```python
pd.merge(df_customer, df_order, on=['id'])
```
