```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('Day',fontsize=10)
plt.ylabel('Temperature',fontsize=10)
plt.title('Weather Lnfo')
```

```python
plt.xlabel('Day',fontsize=10)
plt.ylabel('Temperature',fontsize=10)
plt.title('Weather Lnfo')

plt.plot(days,max_t)
```

```python
plt.xlabel('Day',fontsize=10)
plt.ylabel('Temperature',fontsize=10)
plt.title('Weather Lnfo')

plt.plot(days,max_t)
plt.plot(days,min_t)
```

```python
plt.xlabel('Day',fontsize=10)
plt.ylabel('Temperature',fontsize=10)
plt.title('Weather Lnfo')

plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
plt.xlabel('Day',fontsize=10)
plt.ylabel('Temperature',fontsize=10)
plt.title('Weather Lnfo')

plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
```

```python
plt.xlabel('Day',fontsize=10)
plt.ylabel('Temperature',fontsize=10)
plt.title('Weather Lnfo')

plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
plt.legend()                        # legend used to indicate each label
plt.show()
```

```python
plt.xlabel('Day',fontsize=10)
plt.ylabel('Temperature',fontsize=10)
plt.title('Weather Lnfo')

plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
plt.legend(loc='best')                    #legend will move in all side,when best is given it will be in best place
plt.show()
```

```python
plt.xlabel('Day',fontsize=10)
plt.ylabel('Temperature',fontsize=10)
plt.title('Weather Lnfo')

plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
plt.legend(loc='best',shadow=True,fontsize="small")                    
plt.show()
```

```python
plt.xlabel('Day',fontsize=10)
plt.ylabel('Temperature',fontsize=10)
plt.title('Weather Lnfo')

plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
plt.legend(loc='best',shadow=True,fontsize="small") 
plt.grid()                                          #from grid we can observe value more easily
plt.show()
```

```python
# Simple bar chart showing revenues of major US tech companies

company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
plt.bar(company,revenue)
plt.show()
```

```python
plt.xlabel("Major Companies")
plt.ylabel("Revenue")
plt.bar(company,revenue)
plt.show()
```

```python
plt.xlabel("Major Companies")
plt.ylabel("Revenue")
plt.title("Revenue Model for Top Companies")
plt.bar(company,revenue)
plt.show()
```

```python
plt.xlabel("Major Companies")
plt.ylabel("Revenue")
plt.title("Profit Model for Top Companies")
plt.bar(company,profit)
plt.show()
```

```python
plt.xlabel("Major Companies")
plt.ylabel("Revenue")
plt.title("Revenue Model for Top Companies")
plt.bar(company,revenue)
plt.bar(company,profit)
plt.show()
```

```python
plt.xlabel("Major Companies")
plt.ylabel("Revenue")
plt.title("Revenue Model for Top Companies")
plt.bar(company,revenue,label="Revenue")
plt.bar(company,profit,label="Profit")
plt.legend()
plt.show()
```

```python
plt.xticks(xpos,company)
plt.xlabel("Company Name")
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.bar(xpos,revenue,label="Revenue")
plt.bar(xpos,profit,label="profit")
plt.legend()
```

```python
plt.xticks(xpos,company)
plt.xlabel("Company Name")
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.bar(xpos-0.2,revenue,label="Revenue")
plt.bar(xpos+0.2,profit,label="profit")
plt.legend()
```

```python
plt.xticks(xpos,company)
plt.xlabel("Company Name")
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="profit")
plt.legend()
```
